```python
import pandas as pd
import seaborn as sns
import plotly.express as px
import matplotlib.pyplot as plt

# Load dataset
df = pd.read_csv("winequality-red.csv", sep=";")  # Upload or use: https://archive.ics.uci.edu/ml/machine-learning-databases/wine-quality/winequality-red.csv
df.head()

# Basic stats
print("Dataset Info:")
print(df.info())
print("\nSummary Statistics:")
print(df.describe())
print("\nMissing Values:")
print(df.isnull().sum())

# Correlation heatmap
plt.figure(figsize=(10, 8))
sns.heatmap(df.corr(), annot=True, cmap="coolwarm", fmt=".2f")
plt.title("Correlation Matrix of Wine Quality Features")
plt.show()

# Distribution of quality
fig = px.histogram(df, x="quality", title="Wine Quality Distribution", nbins=10)
fig.update_layout(bargap=0.2)
fig.show()

# Scatter plot: Alcohol vs. Quality
fig = px.scatter(df, x="alcohol", y="quality", title="Alcohol vs. Wine Quality",
                 labels={"alcohol": "Alcohol Content", "quality": "Wine Quality"})
fig.show()

# Box plot: Quality vs. pH
fig = px.box(df, x="quality", y="pH", title="pH Distribution by Wine Quality")
fig.show()
```

```python
from sklearn.preprocessing import StandardScaler

# Features and target
X = df.drop("quality", axis=1)
y = df["quality"]

# Scale features
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)
X_scaled_df = pd.DataFrame(X_scaled, columns=X.columns)
X_scaled_df["quality"] = y

# Save processed data
X_scaled_df.to_csv("winequality-red-processed.csv", index=False)

# Download from Colab
from google.colab import files
files.download("winequality-red-processed.csv")
```
